```python
import numpy as np
import matplotlib.pyplot as plt
```

### P1. 

```python
imge =np.array([[0,100,0,0,0,100], 
                [100,0,100,0,100,0],
                [0,0,0,100,0,0],
                [100,0,100,0,100,0],
                [0,100,0,0,0,100]  
                ], dtype=np.uint8              
               )
imge.shape

```

```python
# A heatmap of imge:
plt.imshow(imge, cmap='gray', interpolation='nearest')
plt.show()
```

```python
#Kernel:
K=np.array([[1,-1,1],[-1,1,-1],[1,-1,1]])
#K=np.array([[0,0,0],[0,1,0],[0,0,0]]) #identity kernel
K
```

```python
def conv(imge, K):
    imge_result = np.zeros(imge.shape)
    for i in range(1,imge.shape[0]-1):
        for j in range(1,imge.shape[1]-1):
            imge_result[i,j]=np.sum(imge[i-1:i+2,j-1:j+2]*K)

    return imge_result[1:imge.shape[0]-1, 1:imge.shape[1]-1]
```

```python
imge_conv = conv(imge,K)
imge_conv
```

```python
# A heatmap of imge in grey scale:
plt.imshow(imge_conv, cmap='gray', interpolation='nearest')
plt.show()
```

```python
# A heatmap of imge:
plt.imshow(imge2, cmap='gray', interpolation='nearest')
plt.show()
```

```python
#Convoution:
imge_2_conv = conv(imge2,K)
#imge_2_conv
```

```python
t=1000
imge_2_umbral = np.where(imge_2_conv>t, 255, 0)
imge_2_umbral.shape
```

```python
def conv_median(imge, K):
    imge_result = np.zeros(imge.shape)
    for i in range(1,imge.shape[0]-1):
        for j in range(1,imge.shape[1]-1):
            imge_result[i,j]=np.median(imge[i-1:i+2,j-1:j+2]*K)

    return imge_result[1:imge.shape[0]-1, 1:imge.shape[1]-1]
```

```python
imge2_conv_median = conv_median(imge2,K)
imge2_conv_median.shape
```

```python
# A heatmap of imge in grey scale:
plt.imshow(imge2_conv_median, cmap='gray', interpolation='nearest')
plt.show()
```
